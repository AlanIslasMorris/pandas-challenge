# PyCity Schools Analysis

Analysis

When assessing the correlation between school budget per capita and the reading and math scores. The correlation is low. But it is very high when assessing correlation between reading and math results. This implies that schools that tend to have a higher mean Math Score also tend to have a high Reading Score.

Using the box plot with the following columns: % Passing Math (%M), '% Passing Reading' (%R), '% Overall Passing' (%OP), 'Average Math Score'(MS), 'Average Reading Score'(RS). Despite we have an n of 15 schools. It is possible to observe that the spread gets wider in the following order %OP>%M>%R>MS>RS. That represents, as it is possible to see in the Facts per school table, that for %OP results per school go from 52 up to 91% the range and the variability is greater than in the case of Reading Score, which ranges from 80 to 84. This represents that regardless of the school most of the students sampled in the district will have a good reading score. But the probability of having a good reading and math scores is lower. That cause, specially in some schools, the overall passing to go down significantly.


Reading and math scores by grade

        It was surprising that the distribution of math grades for students from 9th up to 12th grade was quite similar. The statistical summary considering mean, max, min, quartiles and std suggests that there are no big differences. It was confirmed using the box plot. Ranges are quite narrow, they go from 7.6 up to 8.4 regardless of the grade.
        I expected that as students advance to different grades the percentage of students with better or worse grades will change but it did not.
        
        In the case of reading results the ranges are even narrower as they go from 80 up to 85 regardless of the grade.
    
Budget per school

     Another interesting finding is that the relationship between the budget per student and the overall passing is opposite of what we expected. It was expected that higher budget per student would be associated with better overall passing, it did not. Actually, reading and math results have a negative correlation with budget per student.  Possibly exams and projects are harder, which leads to a lower porcentage of overall passing. Another possibility would be that students are less engaged in their courses or perhaps the quality of those schools is not as good as in the others.
    The budget per student in the 15 schools ranges from $248,087.0 to $3,124,928. 
    The mean budget per student is $1,643,295.2.
  
    
Type of school

    When we make the breakdown by type of school results are better for charters. In math, reading and overall passing rates results were higher in charter schools.
    
Scores by school size

    Small and medium schools had better scores than large schools

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = school_data_complete.school_name.unique()
school_count = school_data_complete[['school_name']].drop_duplicates().count()[0]
school_count
```

```python
# Calculate the total number of students
student_count = school_data_complete.student_name.count()
student_count_ = school_data_complete[['student_name', 'school_name']].drop_duplicates().count()[0]
formatted_student_count = "{:,d}".format(student_count)
print(formatted_student_count)
```

```python
# This will not take into account students with the same name that go to different schools
student_count_ = school_data_complete[['student_name', 'school_name']].drop_duplicates().count()[0]
formatted_student_count_ = "{:,d}".format(student_count_)
print(formatted_student_count_)
```

```python
# Calculate the total budget
total_budget = school_data_complete.budget.sum()
formatted_budget = "{:,d}".format(total_budget)
print(formatted_budget)
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete.math_score.mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete.reading_score.mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage =passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & 
    (school_data_complete["reading_score"] >= 70)
                                                ].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({
    "Total Students": [student_count_],
    "Total Budget": [total_budget],
    "No of schools":[school_count],
    "Avg math score":[average_math_score],
    "Avg reading score":[average_reading_score],
    "% passing math ":[passing_math_percentage],
    "% passing reading":[passing_reading_percentage],
    "Overall passing rate":[overall_passing_rate]
})

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Melt the district_summary DataFrame
melted_df = pd.melt(district_summary)

# Rename the columns
melted_df.columns = ['Attribute', 'Value']

melted_df
```

## School Summary

```python
# Use the code provided to select all of the school types
school_types = school_data_complete['type'].unique()
# Calculate the total student count per school
school_types = (school_data_complete.groupby(['school_name', 'type'])
                                     .count()
                                     .reset_index()
                )

school_types = school_types[['school_name', 'type']]
school_types.head(2)
```

```python
# Calculate the total student count per school
students_per_school_counts = (school_data_complete.groupby(
                                    ['school_name'])['Student ID']
                                     .count()
                                     .sort_values(ascending = False)
                                     .reset_index(name='Total Students')
                  )
students_per_school_counts.head(2)
```

```python
# Calculate the total school budget and per capita spending per school
budget_per_school = (school_data_complete.groupby(
                                    ['school_name'])['budget']
                                    .sum()
                                    .sort_values(ascending = False)
                                    .reset_index(name='budget')
                  )
budget_per_school.head(2)
```

```python
budget_per_school = pd.merge(students_per_school_counts,budget_per_school, on = 'school_name')
budget_per_school.head(2)
```

```python
# Calculate the per capita spending for each school
budget_per_school['budget_per_cap per_school'] = budget_per_school['budget'] / budget_per_school['Total Students']
budget_per_school.head(2)
```

```python
# Calculate the average test scores per school
per_school_math = (school_data_complete.groupby(
                                    ['school_name'])['math_score']
                                    .mean()
                                    .sort_values(ascending = False)
                                    .reset_index(name = 'math_score')
                  )

```

```python
per_school_reading = (school_data_complete.groupby(
                                    ['school_name'])['reading_score']
                                    .mean()
                                    .sort_values(ascending = False)
                                    .reset_index()
                  )
```

```python
# Calculate the number of students per school with math scores of 70 or higher
students_passing_math = school_data_complete[school_data_complete['math_score'] >= 70]
school_students_passing_math = (students_passing_math.groupby(
                                    ['school_name'])['Student ID']
                                    .count()
                                    .sort_values(ascending = False)
                                    .reset_index(name = 'N passing math')
                  )
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
students_passing_reading = school_data_complete[school_data_complete['reading_score'] >= 70]
school_students_passing_reading = (students_passing_reading.groupby(
                                    ['school_name'])['Student ID']
                                    .count()
                                    .sort_values(ascending = False)
                                    .reset_index(name = 'n passing reading')
                  ) 
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size().reset_index(name = 'n passing read & math')
```

```python
# Merge the dataframes using the 'school_name' column as the key
per_school_summary = pd.merge(budget_per_school,per_school_math, on = 'school_name', how = 'left')
per_school_summary = pd.merge(per_school_summary,per_school_reading, on = 'school_name', how = 'left')
per_school_summary = pd.merge(per_school_summary,school_students_passing_math, on = 'school_name', how = 'left')
per_school_summary = pd.merge(per_school_summary,school_students_passing_reading, on = 'school_name', how = 'left')
per_school_summary = pd.merge(per_school_summary,school_students_passing_math_and_reading, on = 'school_name', how = 'left')
per_school_summary['per_school_passing_math'] = per_school_summary['N passing math'] / per_school_summary['Total Students'] * 100
per_school_summary['per_school_passing_reading'] = per_school_summary['n passing reading'] / per_school_summary['Total Students'] * 100
per_school_summary['overall_passing_rate'] = per_school_summary['n passing read & math'] / per_school_summary['Total Students'] * 100
```

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by `school_name` and take the mean of the `math_score` column for each.
ninth_grade_math_scores = (ninth_graders.groupby(
                                    ['school_name'])['math_score']
                                    .mean()
                                    .sort_values(ascending = False)
                                    .reset_index()
                          )
ninth_grade_math_scores = ninth_grade_math_scores.rename(columns={'math_score': 'math_score 9th'})

tenth_grader_math_scores = (tenth_graders.groupby(
                                    ['school_name'])['math_score']
                                    .mean()
                                    .sort_values(ascending = False)
                                    .reset_index()
                          )
tenth_grader_math_scores = tenth_grader_math_scores.rename(columns={'math_score': 'math_score 10th'})

eleventh_grader_math_scores = (eleventh_graders.groupby(
                                    ['school_name'])['math_score']
                                    .mean()
                                    .sort_values(ascending = False)
                                    .reset_index()
                          )
eleventh_grader_math_scores = eleventh_grader_math_scores.rename(columns={'math_score': 'math_score 11th'})

twelfth_grader_math_scores = (twelfth_graders.groupby(
                                    ['school_name'])['math_score']
                                    .mean()
                                    .sort_values(ascending = False)
                                    .reset_index()
                          )
twelfth_grader_math_scores = twelfth_grader_math_scores.rename(columns={'math_score': 'math_score 12th'})


# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.merge(ninth_grade_math_scores,tenth_grader_math_scores, on = 'school_name', how = 'left')
math_scores_by_grade = pd.merge(math_scores_by_grade,eleventh_grader_math_scores, on = 'school_name', how = 'left')
math_scores_by_grade = pd.merge(math_scores_by_grade,twelfth_grader_math_scores, on = 'school_name', how = 'left')

# Minor data wrangling
math_scores_by_grade.index.name = None

# Display the DataFrame
math_scores_by_grade
```
